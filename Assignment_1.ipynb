```python
A = 100
B = 5
count = 0
temp_A = A

if A % B == 0:
    while temp_A % B == 0 and temp_A != 0:
        temp_A = temp_A / B
        count += 1
    print(f"{A} is divisible by {B} and can be divided {count} times.")
else:
    print(f"{A} is not purely divisible by {B}.")
```

```python
numbers = list(range(1, 26)) # List from 1 to 25

for num in numbers:
    if num % 3 == 0:
        print(f"{num} is divisible by 3")
    else:
        print(f"{num} is NOT divisible by 3")
```
